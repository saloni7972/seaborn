Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ANSWER==1)Line plot=A Line Plot Is The Simplest Plot In All Plotting Types, As It Is The Visualization Of A Single Function
2)Bar plot=A Bar Chart Is A Way Of Comparing A Set Of Categorical Data. It Is Better To Convert Continuous Data To Bins Before Plotting. 
3)Scatter plot=Scatter Plot Is The Same As A Line Plot, In A Line Plot Instead Of Points Being Joined By Line Segments, The Points Are Shown Individually With A Dot, Circle, Or Any Other Shape
4)Histogram=A Histogram Is A Graph That Shows The Underlying Frequency Distribution Of A Set Of Continuous Data
5)Pair plot=Pair Plot Creates A Grid Of Axis Such That Each Numeric Variable In Data Will Create A Plot Between Each Other The Y-Axis Across A Single Row And The X-Axis Across A Single Column.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
```

```python
fmri=sns.load_dataset('fmri')
```

```python
fmri
```

```python
sns.lineplot(x=fmri.timepoint,y=fmri.signal)
```
